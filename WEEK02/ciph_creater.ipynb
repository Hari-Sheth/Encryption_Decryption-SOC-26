```python
import numpy as np
import re
```

```python
with open("shakespeare.txt", "r", encoding="utf-8") as f:
    all_text = f.read()
all_text = all_text.lower()
all_text = re.sub(r'[^a-z\s]', '', all_text)
all_text = list(all_text)

for i in range(len(all_text)):
    if all_text[i] == ' ': 
        all_text[i] = (ord(all_text[i])-32+26)
    else:
        all_text[i] = (ord(all_text[i])-97)

all_text = all_text[:-5000074]
print(len(all_text))

bricks = [all_text[i:i+50] for i in range(0, len(all_text), 50)]
print(len(bricks))
print(bricks[101])
bricks = np.array(bricks)
```

```python
def shift(x, i):
    return (x+i)%27
```

```python
ciph = (bricks[None, :, :] + np.arange(27)[:, None, None]) % 27
```

```python
ciph = ciph.astype(np.uint8)
np.save('cipher.npy', ciph)
```
