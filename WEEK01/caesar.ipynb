```python
symbols = (
    ' ', 'e', 't', 'a', 'o', 'i', 'n', 's', 'h', 'r',
    'd', 'l', 'c', 'u', 'm', 'w', 'f', 'g', 'y', 'p',
    'b', 'v', 'k', 'j', 'x', 'q', 'z',
    '.', ',', ';', ':', '/', '@'
)
```

```python
text1 = 'dyvkrwwrzbcksvz:xkdyeck@bv@rbvukr:ukw/btvckz:kbvruz:vccks/dykgzdy/edkr:ukgzdyz:kdyvkbvr,;mkdyv:kcyr,,kzdksvkdz;vkd/kcvdkdyvkczhkxv:d,v;v:kd/kg/b.nkdr.z:xk/buvbmke@/:kdyvkrtt/;@,zcyz:xk/wkdyvzbkuvczx:mkzk;riksvkceuuv:,ikdbr:c@/bdvuk/edk/wkdyzck@,rtvmkr:ukdyrdkr,,ki/ebkw/btvckz:kdyvkcr;vkdz;vksvk/:kdyvkwzv,ukd/k;vvdk;vl'
text2 = 'zkg/e,uksvkx,rukd/k.:/gkdyvk:r;vckr:ukaer,zdzvck/wkdyvkczhkxv:d,v;v:kgyztykrbvkd/krtt/;@,zcykdyvkuvczx:;v:dnkw/bkdyrdkzdk;riksvkzkcyr,,ksvkrs,vmke@/:k.:/g,vuxvk/wkdyvk@rbdzvcmkd/kxzfvki/ekc/;vkwebdyvbkrufztvk:vtvccrbikd/ksvkw/,,/gvukdyvbvz:'
```

```python
def find_freq(text: str):
    freq_map = {}
    for t in symbols:
        freq_map[t] = freq_map.get(t, 0)
    for t in text:
        freq_map[t] = freq_map.get(t, 0) + 1
    freq_map = dict(sorted(freq_map.items(), key=lambda x: x[1], reverse=True))
    return freq_map
```

```python
print(find_freq(text1))
```

```python
def find_cipher(text: str):
    return dict(zip(find_freq(text).keys(), symbols))
print(find_cipher(text1))
```

```python
def get_msg(text: str, cipher = None):
    cipher = find_cipher(text) if cipher == None else cipher
    s = ''
    for t in text:
        s += cipher[t]
    return s
```

```python
print(get_msg(text1))
print(get_msg(text2))
```

We can see that the first word is 'tse' for text1, the most probable meaning is 'the', hence we make the key of 'y' as 'h'

```python
cipher = find_cipher(text1)
cipher['y'], cipher['c'] = cipher['c'], cipher['y']
```

```python
print(get_msg(text1, cipher))
```

Now we are kind of sure that the letters 't', 'h' and 'e' are decoded correctly, looking at the word 'theo' we can guess that the in place of 'o' it should be 'm', 'n' or 'y'. Also, we see a word 'tn', which is obviously 'to', hence 'n' should have been 'o'

```python
cipher['/'], cipher[':'] = cipher[':'], cipher['/']
print(get_msg(text1, cipher))
```

Now we can see a phrase 'on the', therefore all of these 5 letters might have been decoded correctly. Now the word 'thit' must obviously be 'that', so we swap again.

```python
cipher['r'], cipher['z'] = cipher['z'], cipher['r']
print(get_msg(text1, cipher))
```

The message has started to take shape, as we see words 'shall', 'their', 'readiness', 'then'. Now seeing 'vithomt' and 'vithin' we can make another swap.

```python
cipher['@'], cipher['g'] = cipher['g'], cipher['@']
cipher['e'], cipher['w'] = cipher['w'], cipher['e']
print(get_msg(text1, cipher))
```

Now again we can try to make the words like 'suddenly' and 'transported'

```python
cipher['x'], cipher['i'] = cipher['i'], cipher['x']
cipher['m'], cipher['@'] = cipher['@'], cipher['m']
print(get_msg(text1, cipher))
```

```python
cipher[';'], cipher['w'] = cipher['w'], cipher[';']
cipher['s'], cipher['x'] = cipher['x'], cipher['s']
print(get_msg(text1, cipher))
```

```python
cipher['w'], cipher['t'] = cipher['t'], cipher['w']
print(get_msg(text1, cipher))
```

```python
cipher['m'], cipher['o'] = cipher['o'], cipher['m']
cipher['f'], cipher['n'] = cipher['n'], cipher['f']
cipher['a'], cipher['l'] = cipher['l'], cipher['a']
print(get_msg(text1, cipher))
```
