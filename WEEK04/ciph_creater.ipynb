```python
with open("shakespeare.txt", "r", encoding="utf-8") as f:
    all_text = f.read()
all_text = all_text.lower()
all_text = re.sub(r'[^a-z\s]', '', all_text)
all_text = list(all_text)

for i in range(len(all_text)):
    if all_text[i] == ' ': 
        all_text[i] = (ord(all_text[i])-32)
    else:
        all_text[i] = (ord(all_text[i])-97+1)

all_text = all_text[:-2000074+2000000]
print(len(all_text))

bricks = [all_text[i:i+100] for i in range(0, len(all_text), 100)]
print(len(bricks))
print(bricks[101])
bricks = np.array(bricks)
```

```python
def make_key ():
    alph = list('abcdefghijklemnopqrstuvwxyz')
    length = random.randint(3,5)
    word = ''.join(random.choices(alph, k=length))
    for i in range(5-length): word+=' '
    return (word, length)

print(make_key())
```

```python
def apply_vignere(s, tup):
    if not tup:    
        word, length = make_key()
    l = []
    word, length = tup
    for i, c in enumerate(s):
        # Replace corrupt values with 0
        if not (0 <= c <= 26):
            c = 0
        if c == 0:
            l.append(0)
        else:
            shift = ord(word[i % length]) - ord('a')
            new_c = (c - 1 + shift) % 26 + 1
            l.append(new_c)

    return np.array(l)
    
```

```python
def list_to_sentence (l):
    s = ''
    for i in l:
        if 1<=i<=26:
            s += chr(i+ord('a')-1)
        else: s += ' '
    return s

def sentence_to_list (s):
    l = []
    for c in s:
        if c == ' ': l.append(0)
        else: l.append(ord(c)-ord('a')+1)
    return l
```

```python
print()
print(list_to_sentence(apply_vignere(bricks[77], ('abcde', 5))))
print(list_to_sentence(bricks[77]))
```

```python
tikitiki = []
for i in range(len(bricks)):
    tikitiki.append(make_key())
print(tikitiki)

x = np.array([apply_vignere(bricks[i], tikitiki[i]) for i in range(len(bricks))])

print(x.shape)

y = np.array([sentence_to_list(tikitiki[i][0]) for i in range(len(tikitiki))])

print(y)


```
