```python
import numpy as np
import re
import random
```

```python
with open("shakespeare.txt", "r", encoding="utf-8") as f:
    all_text = f.read()
all_text = all_text.lower()
all_text = re.sub(r'[^a-z\s]', '', all_text)
all_text = list(all_text)

for i in range(len(all_text)):
    if all_text[i] == ' ': 
        all_text[i] = (ord(all_text[i])-32+26)
    else:
        all_text[i] = (ord(all_text[i])-97)

all_text = all_text[:-5000074]
print(len(all_text))

bricks = [all_text[i:i+50] for i in range(0, len(all_text), 50)]
print(len(bricks))
print(bricks[101])
bricks = np.array(bricks)
```

```python
def shift(x, i):
    return (x+i)%27
```

```python
# Define 3 fixed substitution ciphers
def make_cipher():
    indices = list(range(27))       # 0-25 = letters, 26 = space
    shuffled = indices.copy()
    np.random.shuffle(shuffled)
    return dict(zip(indices, shuffled))

def apply_cipher(sentence, cipher):
    return [cipher.get(ch, ch) for ch in sentence]  # returns list of ints, not string

CIPHERS = [make_cipher() for _ in range(10)]

# Build dataset — each sample is (ciphered_sentence, original_sentence)
MAX_LEN = 200
all_samples = []

for sentence in bricks:
    if not (5 <= len(sentence) <= MAX_LEN):
        continue
    for cipher in CIPHERS:
        ciphered = apply_cipher(sentence, cipher)
        all_samples.append((np.array(ciphered), np.array(sentence)))

random.shuffle(all_samples)

all_samples = np.array(all_samples)
print(all_samples.shape)
```

```python
ciph = all_samples.astype(np.uint8)
np.save('cipher.npy', ciph)
```
